```python
import numpy as np
from PIL import Image
import cv2
import math 
import skimage
from skimage import io, restoration,data
from matplotlib import pyplot as plt
image = data.camera()
```

```python
def reduce_image_intensity(image,intensity_levels):
    deivisons= intensity_levels
    steps = math.ceil(math.log(deivisons,2))
    modified_steps= 2**steps
    print("actual steps to be modifeid : ",modified_steps)
    if(modified_steps==2):
        image2=np.around(image/256,0)*256
    else:
        image2=np.around(image/(256/modified_steps),0)*(256/modified_steps)
    return image2
```

```python
fig,arr= plt.subplots(1,2)
fig.suptitle("General Title")
arr[0].imshow(image,cmap='gray')
arr[1].imshow(reduce_image_intensity(image,250),cmap='gray')

```

```python


# Get the image dimensions
height, width = image.shape[:2]

# Create a rotation matrix
rotation_matrix_45 = cv2.getRotationMatrix2D((width/2, height/2), 45, 1)
rotation_matrix_90 = cv2.getRotationMatrix2D((width/2, height/2), 90, 1)


# Rotate the image
rotated_img_45 = cv2.warpAffine(image, rotation_matrix_45, (width, height))
rotated_img_90 = cv2.warpAffine(image, rotation_matrix_90, (width, height))
```

```python
fig,arr= plt.subplots(1,2)
fig.subplots_adjust(wspace=0.5) # increase the horizontal space between subplots to 0.5
arr[0].imshow(rotated_img_45)
arr[1].imshow(rotated_img_90)
```

```python
import numpy as np
image = cv2.imread('car.jpg')
height, width = image.shape[:2]
block_size = 3
averaged_image_3_3 = image.copy()
for i in range(0, height-block_size+1, block_size):
    for j in range(0, width-block_size+1, block_size):
        block = image[i:i+block_size, j:j+block_size]
        avg = block.mean()
        averaged_image_3_3[i:i+block_size, j:j+block_size] = avg
fig,arr= plt.subplots(1,2)
fig.subplots_adjust(wspace=0.5) # increase the horizontal space between subplots to 0.5
arr[0].imshow(image)
arr[1].imshow(averaged_image_3_3)
```
